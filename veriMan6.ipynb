# Gözlem ve Değişken Seçimi loc&iloc 

```python
import numpy as np
```

```python
import pandas as pd
```

```python
m=np.random.randint(1,30,size=(10,3))
df=pd.DataFrame(m,columns=["var1","var2","var3"])
df
```

```python
#loc: Tanımlandığı,isimlendirildiği şekli ile seçim yapılır
```

```python
df.loc[0:3] #Gözlem birimlerini isimlere göre getirir (a,b,c,d) yapsaydık a:d derdik mesela
```

```python
df.loc[0:3,"var1":"var2"] #Sütun da aynı şekilde 
```

```python
#iloc: İsmi önemsemeden klasik indexe göre getirir
```

```python
df.iloc[0:3] #a,b,c,d olsaydı da 0:3 yapınca yine aynı gözlem birimleri gelirdi
```

```python
df.iloc[0:3]["var1"] #Bu şekilde select işlemi yapılabilir
```

```python
df.iloc[0:3][["var1","var2"]]  #Fancy kullanımı
```

```python
df.iloc[0,0]
```

```python
df.iloc[0,0]
```

# Koşullu Eleman İşlemleri

```python
df
```

```python
df[(df.var1>15) & (df.var2<18)]
```

```python
df.loc[(df.var1>15) & (df.var2<18),["var1","var2"]] #loc ile df içerisine de sütun bilgisi verebiliyoruz
```

```python
df[(df.var1>15) & (df.var2<18)][["var1","var2"]] #Fancy ile variable selection da yapılabilir
```

```python
#İleri birleştirme işlemleri
```

```python
df1=pd.DataFrame({'calisanlar':["Ali","Veli","Ayse","Fatma"],
                  'grup':["Muhasebe","Muhendislik","Muhendislik","İK"]  })
df1
```

```python
df2=pd.DataFrame({'calisanlar':["Ali","Veli","Ayse","Fatma"],
                  'ilk_giris':[2010,2009,2014,2019]})
df2
```

```python
df3=pd.merge(df1,df2)
df3
```

```python
pd.merge(df1,df2,on="calisanlar") #Yukarıda oto calisanlar uzerinde birlestirmisti parametre de alıyor.
```

```python
df4=pd.DataFrame({'grup':['Muhasebe','Muhendislik','İK'],
                  'mudur':['Ahmet','Mehmet','Mustafa']})
df4
```

```python
#ManytoOne
```

```python
pd.merge(df3,df4)
```

```python
#Many to Many
```

```python
df5=pd.DataFrame({'grup':['Muhasebe','Muhasebe','Muhendislik','Muhendislik','İK','İK'],
                  'yetenekler':['matematik','excel','kodlama','linux','excel','yonetim']})
df5
```

```python
pd.merge(df3,df5)
```
