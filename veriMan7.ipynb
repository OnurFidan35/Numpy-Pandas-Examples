# İleri Toplulaştırma İşlemleri (Aggregate, filter, transform, apply)

```python
df=pd.DataFrame({
    'gruplar':["A","B","C","A","B","C"],
    'deg1':[10,11,52,23,43,55],
    'deg2':[40,21,32,83,13,25],
})
df
```

```python
#aggregate
```

```python
df.groupby("gruplar").aggregate(["min",np.median,"max"]) #Min ve max pandas içerisindeki fonksiyonlar iken medianı dışarıdan(numpy) aldık.
#Bu özelliği belirli değişkenlerin belirli istatistiki özelliklerini seçerek almak için kullanıyoruz. 
```

```python
df.groupby("gruplar").aggregate({"deg1":"max","deg2":"min"})
```

```python
df
```

```python
def filter_func(x):
    return x["deg1"].std()>9
```

```python
df.groupby("gruplar").std()
```

```python
df.groupby("gruplar").filter(filter_func)
```

```python
#Transform
```

```python
df
```

```python
df_a=df.iloc[:,1:3] #Kategorik değişken dışını aldık sayısal işlem yapacağız
df_a
```

```python
df_a.transform(lambda x: (x-x.mean())/ x.std()) #Standart dönüştürme işlemi 
#Bu şekilde transform fonksiyonu aracılığıyla gözlem verileri üzerine kendi tanımladığımız fonksiyonu uyguladık
```

```python
#Apply
```

```python
df_a
```

```python
df_a.apply(np.sum)
```

```python
df_a.apply(np.mean)
```

```python
df
```

```python
df.groupby("gruplar").apply(np.sum)
```

# Pivot Tablolar


```python
titanic=sns.load_dataset("titanic")
titanic.head()
```

```python
titanic.groupby("sex")["survived"].mean()
```

```python
titanic.groupby("sex")[["survived"]].mean()
```

```python
titanic.groupby(["sex","class"])[["survived"]].mean()
```

```python
titanic.groupby(["sex","class"])[["survived"]].aggregate("mean")
```

```python
titanic.groupby(["sex","class"])[["survived"]].aggregate("mean").unstack()
```

```python
#Pivot ile daha kolay yoldan table
```

```python
titanic.pivot_table("survived",index="sex",columns="class")
```

```python
titanic.age.dropna().head(10)
```

```python
age=pd.cut(titanic["age"],[0,18,90])
age.dropna().head(10) #Dropna eksik verileri götürdü(denemek için yaptım hoca kullanmamış)
```

```python
titanic.pivot_table("survived",index=["sex",age],columns="class")
```
